```python
#Databáze modelů

import matplotlib.pyplot as plt

# Databáze Škoda modelů
skoda_models = []
    ("Octavia", 210, 1996, 2025),
    ("Fabia", 190, 1999, 2025),
    ("Superb", 250, 2001, 2025),
    ("Kodiaq", 230, 2016, 2025),
    ("Karoq", 210, 2017, 2025),
    ("Scala", 210, 2019, 2025),
    ("Rapid", 200, 2012, 2025),
    ("Enyaq", 180, 2020, 2025)



1)Filtrování seznamu

všechny modely s maximální rychlostí nad 200 km/h:

fast_cars = [model for model in skoda_models if model[1] > 200]
print("Modely s rychlostí nad 200 km/h:", fast_cars)


2)Řazení podle různých atributů

podle rychlosti nebo podle roku první výroby:

# Řazení podle maximální rychlosti (sestupně)
sorted_by_speed = sorted(skoda_models, key=lambda x: x[1], reverse=True)

# Řazení podle roku první výroby (vzestupně)
sorted_by_year = sorted(skoda_models, key=lambda x: x[2])

print("Seřazeno podle rychlosti:", sorted_by_speed)
print("Seřazeno podle roku výroby:", sorted_by_year)


3)Výpočet popisných statistik

průměrná rychlost, nejvyšší a nejnižší rychlost:

speeds = [model[1] for model in skoda_models]
average_speed = sum(speeds) / len(speeds)
max_speed = max(speeds)
min_speed = min(speeds)

print(f"Průměrná rychlost: {average_speed} km/h")
print(f"Nejvyšší rychlost: {max_speed} km/h")
print(f"Nejnižší rychlost: {min_speed} km/h")

4)Grafická reprezentace – histogram

Ukážeme rozdělení maximální rychlosti:

plt.hist(speeds, bins=5, color='skyblue', edgecolor='black')
plt.title("Rozdělení maximální rychlosti modelů Škoda")
plt.xlabel("Maximální rychlost (km/h)")
plt.ylabel("Počet modelů")
plt.show()

```
